```python
# Dependencies
import numpy as np
import datetime as dt
import pandas as pd
#add the specific style we want, fivethirtyeight
from matplotlib import style
style.use('fivethirtyeight')
#plotting framework
import matplotlib.pyplot as plt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station

#references the classes that were mapped in each table.
Base.classes.keys()
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
#resource: https://stackoverflow.com/questions/12019766/how-to-get-month-and-year-from-date-field-in-sqlalchemy
temp_june=session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
```

```python
#  3. Convert the June temperatures to a list.
print(temp_june)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
temp_june_df=pd.DataFrame(temp_june, columns=['date', 'temperature'])
temp_june_df.head(3)
```

```python
temp_june_df.set_index(temp_june_df['date'], inplace=True)
temp_june_df.to_string(index=False)
temp_june_df = temp_june_df.sort_index()
temp_june_df.to_string(index=False)

temp_june_df.plot.hist(bins=20)
plt.title('June Temperature Observations')
plt.tight_layout()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
temp_june_df.describe()
```

```python
# Additional query that filters the precipitation table to retrieve the temperatures for the month of June. 
prcp_june=session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
```

```python
print(prcp_june)
```

```python
prcp_june_df=pd.DataFrame(prcp_june, columns=['date', 'precipitation'])
prcp_june_df.head(3)
```

```python
prcp_june_df.set_index(prcp_june_df['date'], inplace=True)
prcp_june_df.to_string(index=False)
prcp_june_df = prcp_june_df.sort_index()
prcp_june_df.to_string(index=False)

prcp_june_df.plot()
plt.title('June Precipitation Observations')
plt.locator_params(nbins=5)
plt.show()
```

```python
prcp_june_df.describe()
```
